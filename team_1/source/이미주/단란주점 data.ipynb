```python
import pyproj
#x = 185872.179201
#y = 444031.699524
#crs_from = pyproj.CRS("EPSG:5181")  # 현재 좌표의 CRS (Korea 2000 좌표계)
#crs_to = pyproj.CRS("EPSG:4326")  # 목표 좌표의 CRS (WGS84 좌표계)
#transformer = pyproj.Transformer.from_crs(crs_from, crs_to, always_xy=True)
#lon, lat = transformer.transform(x, y)
#print("위도:", lat)
#print("경도:", lon)
#x,y 좌표 변환

lat_value=[]
lng_value=[]
for num in dx_filered.index:
    x = dx_filered['좌표정보(X)'][num]
    y = dx_filered['좌표정보(Y)'][num]
    crs_from = pyproj.CRS("EPSG:5181")  # 현재 좌표의 CRS (Korea 2000 좌표계)
    crs_to = pyproj.CRS("EPSG:4326")  # 목표 좌표의 CRS (WGS84 좌표계)
    transformer = pyproj.Transformer.from_crs(crs_from, crs_to, always_xy=True)
    lon, lat = transformer.transform(x, y)
    lat_value.append(lat)
    lng_value.append(lon)
    
    
dx_filered['위도'] = lat_value 
dx_filered['경도'] = lng_value 
```
