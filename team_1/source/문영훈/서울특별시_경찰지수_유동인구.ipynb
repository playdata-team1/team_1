```python
import pandas as pd
```

```python
df = pd.read_excel('data/원본/상주(야간)·주간인구_20230713151847.xlsx')
```

```python
df
```

```python
df1 = df.loc[3:,['자치구별(2)','2020','2020.7','2020.1']]
df1.rename(columns = {'자치구별(2)':'시군구','2020':'상주인구','2020.7':'주간인구','2020.1':'유입인구'}, inplace = True)
```

```python
df1.reset_index(drop = True)
```

```python
crime = pd.read_excel('data/git_copy/data_21_전처리/범죄/범죄발생_검거_발생율_검거율_2021.xlsx', index_col = 0)
```

```python
crime = crime.loc[:, ['시군구', '범죄발생률']]
```

```python
df2 = pd.merge(df1, crime, how = 'inner', on = '시군구')
```

```python
df3 = df2.loc[:,['시군구','유입인구','범죄발생률']]
df3['유입인구'] = df3['유입인구'].astype(int)
```

```python
df3.corr()
```

```python
# df3.to_excel('data/preprocessing/유입인구_범죄발생율.xlsx')
```

```python
df4 = df1.sort_values('유입인구',  ascending=  False)
```

```python
df4['주간인구 대비 유입인구'] = df4['주간인구']/df4['유입인구']
df4.reset_index(inplace = True, drop = True)
# df4 = df4.iloc[:,1:]
```

```python
df4
```

```python
df5 = pd.merge(df4, crime, how = 'inner', on = '시군구')
```

```python
df5
```

```python
# df4 # 주간인구와 상주인구의 차이가 클 수록 값이 낮게나올 것이다
```

```python
df6 = df5.sort_values('주간인구 대비 유입인구', ascending = True)
```

```python
df6.reset_index(drop = True, inplace = True)
```

```python
df6
```

```python
judge = pd.read_excel('data/preprocessing/최종판별.xlsx', index_col = 0)
```

```python
df7 = pd.merge(df6,judge, how = 'inner', on = '시군구')
```

```python
df7.to_csv('data/preprocessing/찐막/유입인구_투입지수.csv')
```
