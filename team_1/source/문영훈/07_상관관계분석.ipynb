```python
c = pd.read_csv('data/preprocessing/찐막/지역구_년도별_범죄발생.csv', encoding = "utf-8")
h17 = pd.read_csv('data/preprocessing/찐막/지역구별병원시설2017.csv', encoding = "utf-8")
h18 = pd.read_csv('data/preprocessing/찐막/지역구별병원시설2018.csv', encoding = "utf-8")
h19 = pd.read_csv('data/preprocessing/찐막/지역구별병원시설2019.csv', encoding = "utf-8")
```

```python
c17 = c.loc[:,['자치구별(2)','2017']]
c17.rename(columns = {'자치구별(2)':'시군구', '2017' : '2017년범죄'}, inplace = True)

c18 = c.loc[:,['자치구별(2)','2018']]
c18.rename(columns = {'자치구별(2)':'시군구', '2018' : '2018년범죄'}, inplace = True)

c19 = c.loc[:,['자치구별(2)','2019']]
c19.rename(columns = {'자치구별(2)':'시군구', '2019' : '2019년범죄'}, inplace = True)
```

```python
d17 = pd.read_csv('data/preprocessing/찐막/서울특별시구별인구2017.csv', encoding = "utf-8")
d18 = pd.read_csv('data/preprocessing/찐막/서울특별시구별인구2018.csv', encoding = "utf-8")
d19 = pd.read_csv('data/preprocessing/찐막/서울특별시구별인구2019.csv', encoding = "utf-8")
```

```python
d17 = d17.iloc[:,1:]
d18 = d18.iloc[:,1:]
d19 = d19.iloc[:,1:]
```

```python
hc17 = pd.merge(h17,c17, how = 'inner', on = '시군구')
hc18 = pd.merge(h18,c18, how = 'inner', on = '시군구')
hc19 = pd.merge(h19,c19, how = 'inner', on = '시군구')
```

```python
hcd17 = pd.merge(hc17, d17, how = 'inner', on = '시군구')
hcd18 = pd.merge(hc18, d18, how = 'inner', on = '시군구')
hcd19 = pd.merge(hc19, d19, how = 'inner', on = '시군구')
```

```python
d = pd.read_excel('data/preprocessing/찐막/서울시_구별_범죄발생_인구수.xlsx')
```

```python
hcd17['인구대비범죄율'] = hcd17['2017년범죄']/d['2017 인구합계']*100
hcd18['인구대비범죄율'] = hcd18['2018년범죄']/d['2018 인구합계']*100
hcd19['인구대비범죄율'] = hcd19['2019년범죄']/d['2019 인구합계']*100
```

```python
hcd17 = hcd17.iloc[:,[0,1,4]]
hcd18 = hcd18.iloc[:,[0,1,4]]
hcd19 = hcd19.iloc[:,[0,1,4]]
```
