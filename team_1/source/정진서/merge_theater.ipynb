```python
theater_branded_2019 = pd.read_excel('../data/극장_구별_주소_2019.xlsx', index_col = 0)
theater_2019 = pd.read_excel('../data/극장_공연장_구별_주소_2019_중복제거.xlsx', index_col = 0)
```

```python
theater_all_2017_total = theater_all_2017.groupby('시군구').agg('count')
theater_all_2018_total = theater_all_2018.groupby('시군구').agg('count')
theater_all_2019_total = theater_all_2019.groupby('시군구').agg('count')
```

```python
theater_all_2017_total
```

```python
theater_all_2017_total = theater_all_2017_total[['사업장명']] # 나머지 컬럼 drop
theater_all_2017_total.rename(columns = {'사업장명' : '2017'}, inplace = True) # 이름 변경
theater_all_2018_total = theater_all_2018_total[['사업장명']] # 나머지 컬럼 drop
theater_all_2018_total.rename(columns = {'사업장명' : '2018'}, inplace = True) # 이름 변경
theater_all_2019_total = theater_all_2019_total[['사업장명']] # 나머지 컬럼 drop
theater_all_2019_total.rename(columns = {'사업장명' : '2019'}, inplace = True) # 이름 변경
```

```python
theater_all_2017_total.reset_index(inplace = True) # 인덱스 초기화
theater_all_2018_total.reset_index(inplace = True)
theater_all_2019_total.reset_index(inplace = True)
```

```python
theater_all_2017_total.insert(0, '시도', '서울특별시') # 태블로 사용을 위한 시도 컬럼 추가
theater_all_2018_total.insert(0, '시도', '서울특별시')
theater_all_2019_total.insert(0, '시도', '서울특별시')
```

```python
theater_years_info = theater_all_2017_total.merge(theater_all_2018_total, on = '시군구', how='inner').merge(theater_all_2019_total, on = '시군구', how = 'inner')
theater_years_info = theater_years_info[['시도', '시군구', '2017', '2018', '2019']]
theater_years_info
```

```python
theater_years_info['2017~2018'] = (theater_years_info['2017'] - theater_years_info['2018']) * -1
theater_years_info['2018~2019'] = (theater_years_info['2018'] - theater_years_info['2019']) * -1
theater_years_info['2017~2019'] = (theater_years_info['2017'] - theater_years_info['2019']) * -1
```
