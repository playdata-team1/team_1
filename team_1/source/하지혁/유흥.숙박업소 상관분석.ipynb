```python
p_crime = people_crime.drop(['Unnamed: 0'], axis=1)
p_crime.head()
```

```python
p_crime['2017범죄발생률'] = p_crime['2017 범죄발생합계'] / p_crime['2017 인구합계']*100
p_crime['2018범죄발생률'] = p_crime['2018 범죄발생합계'] / p_crime['2018 인구합계']*100
p_crime['2019범죄발생률'] = p_crime['2019 범죄발생합계'] / p_crime['2019 인구합계']*100
p_crime.head()
```

```python
crime = p_crime[['시도', '시군구', '2017범죄발생률', '2018범죄발생률', '2019범죄발생률']]
crime
```

```python
motel_cor = pd.merge(crime, motel, on='시군구')
motel_cor.head()
```

```python
np.corrcoef(motel_cor['2017범죄발생률'],motel_cor['2017숙박업소합계'])
```

```python
np.corrcoef(motel_cor['2018범죄발생률'],motel_cor['2018숙박업소합계'])
```

```python
np.corrcoef(motel_cor['2019범죄발생률'],motel_cor['2019숙박업소합계'])
```

```python
adult_cor = pd.merge(crime, adult, on='시군구')
adult_cor.head()
```
