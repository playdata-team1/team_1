```python
import pandas as pd
import numpy as np
```

```python
police_ = pd.read_excel('/Users/hajihyuk/Documents/1차 미니 프로젝트/합계/서울시_구별_경찰관수.xlsx', index_col=0)
police = police_.rename(columns={'2021년':'경찰관수'})
police.head()
```

```python
people_ = pd.read_csv('/Users/hajihyuk/Documents/1차 미니 프로젝트/합계/2021인구수.csv', index_col=0)
people = people_.rename(columns={'2021년':'인구수'})
people.head()
```

```python
peo_pol = pd.merge(people, police, on = '시군구')
peo_pol.head()
```

```python
add = pd.read_excel('/Users/hajihyuk/Documents/1차 미니 프로젝트/합계/지구대_경찰인력.xlsx', index_col=0)
add.head()
```

```python
peo_police = pd.merge(peo_pol, add, on = '시군구')
peo_police.head()
```

```python
peo_police1 = peo_police[['시군구', '인구수', '경찰관수', '지구대 경찰수']]
peo_police1.head()
```

```python
peo_police1['총경찰관수'] = (peo_police1['경찰관수'] + peo_police1['지구대 경찰수'])
peo_police1.head()
```

```python
people_police = peo_police1[['시군구', '인구수', '총경찰관수']]
people_police.head()
```

```python
people_police['경찰1인당 담당인구수'] = people_police['인구수'] / people_police['총경찰관수']
people_police
```

```python
people_police['경찰1인당 담당인구수'].describe()
```

```python
p1 = people_police[people_police['경찰1인당 담당인구수'] >= 347.026545]
p2 = people_police[people_police['경찰1인당 담당인구수'] < 347.026545]
```

```python
p1['판별'] = True
p1.head()
```

```python
p2['판별'] = False
p2.head()
```

```python
p3 = pd.concat([p1,p2])
```

```python
p3.reset_index(inplace=True)
p4 = p3.rename(columns={'총경찰관수':'경찰관수'})
respon = p4[['시군구', '인구수', '경찰관수', '경찰1인당 담당인구수', '판별']]
respon
```

```python
respon.to_excel('/Users/hajihyuk/Documents/1차 미니 프로젝트/경찰1인담_담당인구수.xlsx')
```
